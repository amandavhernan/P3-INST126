```python
# Reads user log file and stores data in dictionary

def logFile(file):
    logsDict = {}
    for log in file:
        items = log.split()
        date = items[0]
        time = items[1]
        activity = items[2]
        server = items [3]
        user_email = items[4]
        
        if user_email in logsDict:
            if date in logsDict[user_email]:
                logsDict[user_email][date].append((time, activity, server))
            else:
                logsDict[user_email][date] = [(time, activity, server)]
        else:
            logsDict[user_email] = {date: [(time, activity, server)]}
        sorted(logsDict[user_email][date])
    return logsDict

# Reads userlog.log

userlog = open("userlog.log")
file = userlog.readlines()
logsDict = logFile(file)
```

```python
# Suspicious activities report

susActivity = {}
susCount = 0
for user_email in logsDict:
    for date in logsDict[user_email]:
        loginCount = sum(1 for i in logsDict[user_email][date] if i[1] == 'login')
        lateLogin = any(i[1] == 'login' and int(i[0].split(':')[0]) < 5 for i in logsDict[user_email][date])

        if loginCount > 5 or lateLogin:
            susCount += 1
            if user_email in susActivity:
                susActivity[user_email].append(date)
            else:
                susActivity[user_email] = [date]
            sorted(susActivity[user_email])
            
new_file = open("suspicious_report.txt", "x")
new_file.write("==============================\n")
new_file.write(f"=== SUSPICIOUS ACTIVITIES ({str(susCount)} cases) ===\n")
new_file.write("==============================\n")

for user_email in susActivity:
        
    activityCount = len(susActivity[user_email])
    new_file.write(f"{user_email}\t\t{str(activityCount)}\n")
        
    for date in susActivity[user_email]:
        new_file.write(f"\tDATE: [{date}] ---\n")
            
        for activity in logsDict[user_email][date]:
            new_file.write(f"\t\t{activity[0]}\t{activity[2]}\n")
                
new_file.close()
```

```python
# Irresponsible behavior report

irresBehavior = {}
irresCount = 0
for user_email in logsDict:
    for date in logsDict[user_email]:
        loginCount = sum(1 for i in logsDict[user_email][date] if i[1] == 'login')
        logoutCount = sum(1 for i in logsDict[user_email][date] if i[1] == 'logout')

        if loginCount > logoutCount:
            irresCount += 1
            if user_email in irresBehavior:
                irresBehavior[user_email].append(date)
            else:
                irresBehavior[user_email] = [date]
            sorted(irresBehavior[user_email])
            
new_file = open("irresponsible_report.txt", "x")
new_file.write("==============================\n")
new_file.write(f"=== IRRESPONSIBLE BEHAVIOR ({str(irresCount)} cases) ===\n")
new_file.write("==============================\n")

for user_email in susActivity:
        
    activityCount = len(susActivity[user_email])
    new_file.write(f"{user_email}\t\t{str(activityCount)}\n")
        
    for date in susActivity[user_email]:
        new_file.write(f"\tDATE: [{date}] ---\n")
            
        for activity in logsDict[user_email][date]:
            new_file.write(f"\t\t{activity[0]}\t{activity[2]}\n")
                
new_file.close()
```

```python
# System glitch report

glitch = {}
glitchCount = 0
for user_email in logsDict:
    for date in logsDict[user_email]:
        logoutCount = sum(1 for i in logsDict[user_email][date] if i[1] == 'logout')
        loginCount = sum(1 for i in logsDict[user_email][date] if i[1] == 'login')

        if logoutCount > loginCount:
            glitchCount += 1
            if user_email in glitch:
                glitch[user_email].append(date)
            else:
                glitch[user_email] = [date]
            sorted(glitch[user_email])
            
new_file = open("glitch_report.txt", "x")
new_file.write("==============================\n")
new_file.write(f"=== SYSTEM GLITCH ({str(glitchCount)} cases) ===\n")
new_file.write("==============================\n")

for user_email in susActivity:
        
    activityCount = len(susActivity[user_email])
    new_file.write(f"{user_email}\t\t{str(activityCount)}\n")
        
    for date in susActivity[user_email]:
        new_file.write(f"\tDATE: [{date}] ---\n")
            
        for activity in logsDict[user_email][date]:
            new_file.write(f"\t\t{activity[0]}\t{activity[2]}\n")
                
new_file.close()
```

```python
# Domain count report

user_emailList = logsDict.keys()
domainsDict = {}
for user_email in user_emailList:
    emailParts = user_email.split('@')
    domain = emailParts[1]
    if domain in domainsDict:
        domainsDict[domain] += 1
    else:
        domainsDict[domain] = 1

new_file = open("domain_report.txt", "x")
new_file.write("==============================\n")
new_file.write(f"=== DOMAIN COUNT ({str(len(domainsDict))} DOMAINS) ===\n")
new_file.write("==============================\n")

for domain in domainsDict:
    new_file.write(f"{domain}\t\t{domainsDict[domain]}\n")

new_file.close()
```
